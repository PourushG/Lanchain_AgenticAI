### Streaming
Most models can stream their output content while it is being generated. By displayng output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced.

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
load_dotenv()

api_key = os.getenv('GROQ_API')
print(api_key)

model = init_chat_model("groq:qwen/qwen3-32b", api_key = api_key)
```

```python
model.invoke("hello, what's up? how are you doing?")
```

```python
for chunk in model.stream("write a 1000 word paragraphs on the future of AI in healthcare."):
    print(chunk.text, end='')
```

### Batch
Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel

```python
responses = model.batch([
    "what is the capital of India?",
    "Who is the president of USA in 1980?",
    "What is the tallest mountain in the world?"
])

for response in responses:
    print(response.text)
```

```python
# max_concurrency limits how many batch requests are processed in parallel to control resource usage and avoid overload

responses = model.batch([
    "what is the capital of India?",
    "Who is the president of USA in 1980?",
    "What is the tallest mountain in the world?"
],
config={'max_concurrency':1},
)

for response in responses:
    print(response.text)
```
