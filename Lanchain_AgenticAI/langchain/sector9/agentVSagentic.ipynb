## AI AGENTS:
They refers to individual software programs designed to perform specific tasks with a degree of autonomy

Functionality: They handels specific task and have limited autonomy
Decision Making: Uses predefined ruels.

single entity with a focused purpose.

example: simple customer service chatbot with a single task to interact with the company's database based on the user's query.


## AGENTIC AI:
They describes a broader framework where mulitple AI agents can collaborate and make decisions independently to achieve a larger goal, empahszing adaptablitiy and learning from experience rather than just following perdefined rules.

Functionality: they focuses on the overall system with the ability to learn, adapt and make complex decsion.
decsion making: autonomous decision based on the real time data and circumtanses.

Network of agents working together to achieve a goal

example: smart home system where we have mulitple task and each task is handled by seperate agents.
