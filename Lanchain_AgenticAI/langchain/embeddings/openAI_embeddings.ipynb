# 1: Loading

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("/home/dtp202505-u05/Desktop/langchain_agenticAI/langchain/1-langchain/example.txt")
document = loader.load()
```

# 2: Converting to Chunks

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
chunks = splitter.split_documents(document)
chunks
```

# 3: Converting chunks into embeddings and then storing it into vectorDB (chromadb)

```python
from langchain_community.vectorstores import Chroma
vectordb = Chroma.from_documents(chunks, embeddings_1024) # it takes the documents chunks and the embeddng model
```

```python
query = "Not my story"
result = vectordb.similarity_search(query)
print(result)
```
