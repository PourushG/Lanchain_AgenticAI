```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["HF_TOKEN"] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# this is for passing a single string query in text

text = "laksfj;lksjfl;skadjf;laksjf;lasdkjf;laksdfj;laskjfd;laksdjf;lkasdfj"

text_embeddings = embeddings.embed_query(text)

print(text_embeddings)
```

```python
# this for passing a list of texts like documents

documents_embeddings = embeddings.embed_documents([text,"asdfsadfasdfsadfsadfasdfsadfsaf"])

print(documents_embeddings[0]) # embeddings for first document in the list
print(documents_embeddings) # all embeddings for all documents in the list
```
