```python
from pydantic import BaseModel # this is the base class for the pydantic model which performs the data validation

class Person(BaseModel):
    name: str
    age: int
    city: str

# person1 = Person(name="Alice", age=30, city=55) # this will raise a validation error because the city has to be a string but it is provided with a integer value
person1 = Person(name="Alice", age=30, city="abc")
print(person1)  # this will not be printed due to the validation error
```

```python
from dataclasses import dataclass

@dataclass
class Persondataclass():
    name: str
    age: int
    city: str

person2 = Persondataclass(name="Alice", age=30, city="abc")
print(person2)  # this will be printed because dataclasses do not perform validation
# person2 = Persondataclass(name="Alice", age=30, city=55)
# print(person2)  # this will not be printed due to the validation error
```

```python
from typing import Optional

class Employee(BaseModel):
    name: str
    age: int
    city: Optional[str] = None  # Optional field, can be None
```

```python
class Classroom(BaseModel):
    room_number: int
    students: list[str]
    capacity: int

# classroom = Classroom(room_number=101, students=["Alice", "Bob"], capacity=30)
classroom = Classroom(room_number=101, students=("Alice", "Bob"), capacity=30) # this will typecast the tupel to a list
print(classroom)  # this will print the classroom details
```

```python
# Model with nested models

class Address(BaseModel):
    street: str
    city: str
    zip_code: int

class Customer(BaseModel):
    id: int
    name: str
    address: Address

customer = Customer(
    id=1,
    name="John Doe",
    address={"street": "asfasdf", "city": "uuuuff", "zip_code": "88383"}
)

print(customer)
```

### Pydantic fields customizaiton and constraints
The field fucitonin Pydantic enhances model fields beyond basic type hints by allowing you to specify validation rules, default vlaues, aliases, and more.

```python
from pydantic import Field

class item(BaseModel):
    name: str=Field(min_length=3, max_length=50)
    price: float=Field(gt=0, lt=1000)  # price must be greater than 0 and less than 1000
    quantity: int=Field(ge=0) # quantity must be greater than or equal to 0

item1 = item(name="Apple", price=34.34, quantity=33)

print(item1)
```
