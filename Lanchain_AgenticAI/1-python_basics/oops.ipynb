```python
#modeling a bank account

class Bank:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Your account has been debited with {amount} and the updated balance is {self.balance}")
    
    def withdrawal(self, amont):
        self.balance -= amont
        print(f"Your account has been credited with {amont} and the updated balance is {self.balance}")


acc1 = Bank("Pourush")
acc1.deposit(100000000000)
print(acc1.balance)
acc1.withdrawal(100000)
print(acc1.balance)

print()

acc2 = Bank("Abc")
acc2.deposit(10000000)
print(acc2.balance)
acc2.withdrawal(100)
print(acc2.balance)
```

Abstract base classes (ABCs) are used to difne common methods for a group of related objects. They can enforce theat derived classes implement particular methods, promoting consistency acorss different implementations.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC): # ABC is a abstract class which is an empty class the vehicle class is inherited from it.
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "car engine started"
    
class Truck(Vehicle):
    def start_engine(self):
        return "Truch engine started"
    
def start_vehicle(vehicle):
    print(vehicle.start_engine())


car = Car()
truck = Truck()
start_vehicle(car)
start_vehicle(truck)
```
