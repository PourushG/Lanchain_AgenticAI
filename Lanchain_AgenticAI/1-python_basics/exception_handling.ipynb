```python
# try, execpt and else

try: 
    num = int(input("enter a number: "))
    result = 1/num
    # a=b
except Exception as ex:
    print(ex)

else:
    print(f"the answer is {result}") # if no exception has occured then this will executed
```

```python
# try, except, else and finally

try: 
    num = int(input("enter a number: "))
    result = 1/num
    # a=b
except Exception as ex:
    print(ex)

else:
    print(f"the answer is {result}") # if no exception has occured then this will executed

finally:
    print("execution complete") # this will always execute whether an exection occured or not
```
