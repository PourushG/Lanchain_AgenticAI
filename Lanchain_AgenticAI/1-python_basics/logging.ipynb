### Python Logging

Loggin is a crucial aspect of any application, providing a way to track events, errors, and operational infromation. Python's built-in logging module  offeres a flexible framework for emitting log messages from Python programs. In this lesson, we will cover the basics of logging, including how to configure loggin, log levels, and best practices for using logging in Python applicaitons.

```python
import logging

# Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

# These are the different log messages based on the severity level
logging.debug("This is a debug message")
logging.info("This si an info message")
logging.warning("This is a warning message")
logging.error("THis is a error message")
logging.critical("This is a critical message")
```

# Log levels
Python's logging module has several log levels that indicate the severity of the events being logged. The levels, in order of increasing severity, are:
1. DEBUG: Detailed information, typically of interest only when diagnosing problems.
2. INFO: Confirmation that things are working as expected.
3. WARNING: An indication that something unexpected happened, or indicative of some problem in the near future (e.g., 'disk space low'). The software is still functioning as expected.
4. ERROR: Due to a more serious problem, the software has not been able to perform some function.
5. CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.


```python
# Configuring logging 
import logging

logging.basicConfig(
    filename='app.log',  # Log messages will be written to this file
    filemode='w',  # 'w' to overwrite the file, 'a' to append
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',)

# These are the different log messages based on the severity level
logging.debug("This is a debug message")
logging.info("This si an info message")
logging.warning("This is a warning message")
logging.error("THis is a error message")
logging.critical("This is a critical message")

```
