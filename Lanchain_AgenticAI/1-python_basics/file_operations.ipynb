```python
# opening a file

file_path = "/home/dtp202505-u05/Desktop/langchain_agenticAI/temp.txt"

def open_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            print(line.strip())
            # print(line, end="")
```

```python
with open(file_path, 'w') as file: # this will open the file in the write mode but override all the content
    file.write("This the content written using write mode operation\nbut this override the entire content of the file")

open_file(file_path)
```

```python
with open(file_path, 'a') as file:
    file.write("\nThis is the content which is written using the append mode\nwhich will not override the existing content in the file but this will append the new content after it.")

open_file(file_path)
```

```python
file_path = "/home/dtp202505-u05/Desktop/langchain_agenticAI/temp.txt"

list_lines = ["First line\n", "Second line\n", "Third line\n"]

with open(file_path, 'a') as file:
    file.writelines(list_lines)

open_file(file_path)
```

```python
# writing to a binary file

file_path = "/home/dtp202505-u05/Desktop/langchain_agenticAI/temp_binary.bin"

data = b'\x00\x01\x02\x03\x04'

with open(file_path, 'wb') as file:
    file.write(data)

open_file(file_path)
```

```python
# Reading from a source file and writing in a destination file

with open(file_path, 'r') as file:
    content = file.read()

dest_path = "/home/dtp202505-u05/Desktop/langchain_agenticAI/destination.txt"
with open(dest_path, 'w') as file:
    file.write(content)

open_file(dest_path)
```

```python
# Reading from a file and counting the number of lines, words, characters.
def count_text(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)

        return line_count, word_count, char_count
    
line_count,word_count,char_count = count_text(file_path)

print(f"line count: {line_count}")
print(f"word count: {word_count}")
print(f"char count: {char_count}")
```

The w+ mode in python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the files, exists, its content is truncated (i.e, the file is overwritten)

```python
#  Writing and reading operation in a file

with open('w_plus_mode_file.txt', 'w+') as file:
    file.write("This is the sample content from a w+ model created file\n")
    file.write("this is the second line")

    # we need to mode the cursor back to the begining because after printing two lines it is pointing to the second line
    file.seek(0) # we are giving it 0 to move to the 0th index

    # Now, we can start read the file
    content = file.readlines()

    print(content)

```
