Pandas is a powerful data manipulation library in python widely used for data analyssi and data cleaning. It proivdes two primary dta structures: Series and Dataframe. A series is a 1D array like object, while dataframe is a 2d, size-mutable and potentially heterogenous tabluar data structure with labeled axes (rows and columns)

```python
import numpy as np
import pandas as pd
```

```python
#Series: A pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a list or an array, but it has additional features such as labels (index) for each element.
data = np.array([1,2,3,4,5])

series = pd.Series(data) # Creating a Series from a NumPy array
print(series)

dict_data = {'a': 1, 'b': 2, 'c': 3}
series = pd.Series(dict_data)  # Creating a Series from a dictionary
print(series)
```

```python
data = [1,1,1,1,1,1,1]
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

series = pd.Series(data, index = index)  # Creating a Series with a custom index
print(series)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
#handling missing values
df.isnull().any(axis=1) # Check if any row has missing values
df.isnull().any(axis=0)  # Check if any column has missing values
# df.isnull().sum()  # Count of missing values in each column
```

```python
df.fillna(0, inplace=True)  # Fill missing values with 0
# df.dropna(inplace=True)  # Drop rows with any missing values
```

```python
#Replacing the missing values with the mean of the column
df['transacitonsid_fillna_mean'] = df['Transaction ID'].fillna(df['Transaction ID'].mean())
df
```

```python
#renaming a column
df = df.rename(columns={'Product Category': 'Categories'})
df
```

```python
df
```

```python
# Changing the data type of a column
df['Date'].astype('datetime64[ns]')
df['Units Sold'].astype('float64')
df['Unit Price'].astype('int64')
df['Squared Unit prince'] = df['Unit Price'].apply(lambda x: x**2)
df
```

```python
# region wise revenue mean

categories_mean = df.groupby('Region')['Total Revenue'].mean()
print(categories_mean)
```

```python
# region and categories wise revenue mean
region_categories_mean = df.groupby(['Region', 'Categories'])['Total Revenue'].mean()
print(region_categories_mean)
```

```python
#grouped aggreated function
grouped_aggrated = df.groupby(['Region'])['Total Revenue'].agg(['mean', 'sum', 'count'])
print(grouped_aggrated.mean())
```

```python
# Merging and joining Dataframes

df1 = pd.DataFrame({
    'Region': ['North', 'South', 'East'],
    'Total Revenue': [1000, 2000, 1500]
})  
df2 = pd.DataFrame({
    'Region': ['North', 'South', 'West'],
    'Total Revenue': [1200, 1800, 1600]
})
```

```python
pd.merge(df1, df2, on='Region', how='inner') # this will merge the two DataFrames on the 'Region' column using an inner join, which means only the rows with matching 'Region' values in both DataFrames will be included in the result.
```

```python
pd.merge(df1, df2, on='Region', how='outer')  # Merging DataFrames on 'Region' column
```

```python
pd.merge(df1, df2, on='Region', how='left')  # this will merge the two DataFrames on the 'Region' column using a left join, which means all rows from df1 will be included, and matching rows from df2 will be added where available. If there is no match, NaN values will be filled in for the columns from df2.
```

```python
pd.merge(df1, df2, on='Region', how='right')  # this will merge the two DataFrames on the 'Region' column using a right join, which means all rows from df2 will be included, and matching rows from df1 will be added where available. If there is no match, NaN values will be filled in for the columns from df1.
```
